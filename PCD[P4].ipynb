```python
# prompt: please make code for rotation using open cv

import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('gambar_1_obito_rotation.png')

# Define the angle of rotation (in degrees)
angle = 29

# Get the image dimensions
height, width = image.shape[:2]

# Calculate the center of the image
center = (width // 2, height // 2)

# Define the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Perform the rotation
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

# Display the original and rotated images
plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")

plt.show()
```

Reflection

```python
# prompt: please make code for translation using open cv

import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np

# get image from link google image
urllib.request.urlretrieve(
    'https://i.pinimg.com/474x/1c/31/ec/1c31ecb7fb53527ca9b0585d038f6acf.jpg',
    "gambar_1_obito_translation.png")

# Load the image
image = cv2.imread('gambar_1_obito_translation.png')

# Define the translation parameters (x and y shifts)
tx = 75  # Shift 75 pixels to the right
ty = 150   # Shift 150 pixels down

# Create the translation matrix
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# Perform the translation
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

# Display the original and translated images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("Translated Image")

plt.show()
```

```python
# prompt: please make code for reflection using open cv

import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np

# get image from link google image
urllib.request.urlretrieve(
    'https://i.pinimg.com/474x/1c/31/ec/1c31ecb7fb53527ca9b0585d038f6acf.jpg',
    "gambar_1_obito_reflection.png")

# Load the image
image = cv2.imread('gambar_1_obito_reflection.png')

# Flip the image horizontally (reflection)
reflected_image = cv2.flip(image, -1)  # -1 for horizontal flip

# Display the original and reflected images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(reflected_image, cv2.COLOR_BGR2RGB))
plt.title("Reflected Image")

plt.show()
```
